```python
#Approssimativo
%matplotlib inline
import math
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import re

from collections import defaultdict
from scipy.stats.stats import pearsonr

df = pd.read_csv('model.csv') 
```

```python
df['Doors']=['' for i in range(0,df.shape[0])]

for index, row in df.iterrows():
    myList=str(row['SubModel']).split()
    newList=[]
    for index1, item in enumerate(myList): 
        match=re.search('^\dD$',item)
        if(match!=None):
            df.iat[index, 36]=item
            continue
        newList.append(item)
        
    newString=" ".join(x for x in newList)
    df.iat[index, 9]=newString
```

```python
for index, row in df.iterrows():
    myList=str(row['Trim']).split()
    newList=[]
    for index1, item in enumerate(myList):
        if(len(myList)>1 and myList[0]=='3' and myList[1]=='R'):
            df.iat[index, 8]='3R'
```

```python
for index, row in df.iterrows():
    myTrim=row['Trim']
    if(pd.isnull(row['Trim'])):
        continue
    df.iat[index, 8]=myTrim.upper()
```

```python
dicTrim=list(df['Trim'].unique())
for index, row in df.iterrows():
    myList=str(row['SubModel']).split()
    newList=[]
    for index1, item in enumerate(myList):
        if item in dicTrim:
            if pd.isnull(row['Trim']):
                df.iat[index, 8]=item
        
        if(item not in dicTrim):
            newList.append(item)

    newString=" ".join(x for x in newList if x!='')
    df.iat[index, 9]=newString
```

```python
for index, row in df.iterrows():
    myList=str(row['SubModel']).split()
    for index1, item in enumerate(myList):
        match=re.search('^\d\.[0-9]+$',item)
        if(match!=None):
            y='L'
            myList[index1]=item+y
            
        match=re.search('^\d\.$',item)
        if(match!=None):
            y='0L'
            myList[index1]=item+y
            
    newString=" ".join(x for x in myList)
    df.iat[index, 9]=newString
```

```python
for index, row in df.iterrows():
    myListS=str(row['SubModel']).split()
    myListM=str(row['Model']).split()
    newList=[]
    ModelExt=False
    
    for index1, item in enumerate(myListS):
        match=re.search('^\d\.[0-9]+L$',item)
         
        if(match!=None and pd.isnull(row['Model'])):
            df.iat[index, 7]=item
            continue
        
        if(match!=None and pd.notnull(row['Model'])):
            for index1, item2 in enumerate(myListM):
                match2=re.search('^\d\.[0-9]+L$',item)
                if(match2!=None):
                    ModelExt=True
                    continue
        
        if(match!=None and ModelExt==False):
            myListM.append(item)
            newStringM=" ".join(x for x in myListM)
            df.iat[index, 7]=newStringM
        
        if(match==None):
            newList.append(item)     
        
    newString=" ".join(x for x in newList)
    df.iat[index, 9]=newString
```

```python
dic=['CAB','EXT','MINIVAN','CONVERTIBLE','COUPE','UTILITY','WAGON']

for index, row in df.iterrows():
    myList=row['SubModel'].split()
    newListSub=[]
    newListExtra=[]
    for index1, item in enumerate(myList):
        if (index1==0 or (index1==1 and str(item) in dic)):
            newListSub.append(item)
        else:
            newListExtra.append(item)

    newStringSub=" ".join(x for x in newListSub if x!='')
    newStringExtra=" ".join(x for x in newListExtra if x!='')
    
    df.iat[index, 37]=newStringSub
    df.iat[index, 9]=newStringExtra
```

```python
for index, row in df.iterrows():
    myList=row['RealSubModel'].split()
    for index1, item in enumerate(myList):
        if(item=='HATCKBACK'):
            myList[index1]='HATCHBACK'
        
        if(item=='HARTOP'):
            myList[index1]='HARDTOP'
    
    myList=sorted(myList, key=len ,reverse=True)
    newString=" ".join(x for x in myList if x!='')
    df.iat[index, 37]=newString
```
