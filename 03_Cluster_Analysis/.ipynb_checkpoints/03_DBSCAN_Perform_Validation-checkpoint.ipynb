## DBSCAN Perform

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv("Training_DU_Clustering.csv", skipinitialspace=True, sep=',')
df
```

```python
columns_toAnalyze = ['VehicleAge', 'VehOdo', 'MMRAcquisitionAuctionAveragePrice', 'VehBCost', 'WarrantyCost', 'CostOverOdo']

M_Values = df[columns_toAnalyze].values

del columns_toAnalyze
```

```python
dist = pdist(M_Values, 'euclidean')
dist = squareform(dist)
print(dist)
```

```python
k = 5
kth_distances = list()
for d in dist:
    index_kth_distance = np.argsort(d)[k]
    kth_distances.append(d[index_kth_distance])
```

```python
plt.plot(range(0, len(kth_distances)), sorted(kth_distances))
plt.ylabel('dist from %sth neighbor' % k, fontsize=18)
plt.xlabel('sorted distances', fontsize=18)
plt.tick_params(axis='both', labelsize=10)
plt.grid(color='black', linestyle=':', linewidth=0.5)
plt.show()
```

```python
del dist
del index_kth_distance
del kth_distances
del d
```

** - Primo Clustering **

```python
dbscan = DBSCAN(eps=0.2, min_samples=k)
dbscan.fit(M_Values)
```

```python
Labels = dbscan.labels_
```

```python
Labels[:40]  # prime 40 label dei primi 40 punti
```

```python
np.unique(Labels, return_counts=True)
```

*** Performance Index: Silhouette Coefficient ***

```python
Silh_primoClustering = silhouette_score(M_Values, Labels)
print('Silhouette: %s' % Silh_primoClustering)
```

# Considerazioni Finali

la miglior clusterizzazione è quella con indice di validazione più distante dal'indice rispettivo ottenuto da un dataset random.

```python
ValidPrimoClustering = abs(Silh_primoClustering) - abs(Silh_primoClusteringRandom)
ValidSecondoClustering = abs(Silh_secondoClustering) - abs(Silh_secondoClusteringRandom)
ValidTerzoClustering = abs(Silh_terzoClustering) - abs(Silh_terzoClusteringRandom)
ValidQuartoClustering = abs(Silh_quartoClustering) - abs(Silh_quartoClusteringRandom)
ValidQuintoClustering = abs(Silh_quintoClustering) - abs(Silh_quintoClusteringRandom)
ValidSestoClustering = abs(Silh_sestoClustering) - abs(Silh_sestoClusteringRandom)
ValidSettimoClustering = abs(Silh_settimoClustering) - abs(Silh_settimoClusteringRandom)
ValidOttavoClustering = abs(Silh_ottavoClustering) - abs(Silh_ottavoClusteringRandom)
ValidNonoClustering = abs(Silh_nonoClustering) - abs(Silh_nonoClusteringRandom)
ValidDecimoClustering = abs(Silh_decimoClustering) - abs(Silh_decimoClusteringRandom)

print('Primo Clustering: ' + ValidPrimoClustering)
print('Secondo Clustering: ' + ValidSecondoClustering)
print('Terzo Clustering: ' + ValidTerzoClustering)
print('Quarto Clustering: ' + ValidQuartoClustering)
print('Quinto Clustering: ' + ValidQuintoClustering)
print('Sesto Clustering: ' + ValidSestoClustering)
print('Settimo Clustering: ' + ValidSettimoClustering)
print('Ottavo Clustering: ' + ValidOttavoClustering)
print('Nono Clustering: ' + ValidNonoClustering)
print('Decimo Clustering: ' + ValidDecimoClustering)

ValidMigliorClustering = max(ValidPrimoClustering, ValidSecondoClustering, ValidTerzoClustering, ValidQuartoClustering,
                             ValidQuintoClustering, ValidSestoClustering, ValidSettimoClustering, ValidOttavoClustering,
                             ValidNonoClustering, ValidDecimoClustering)

print('la miglior clusterizzazione è quella con indice di validazione distante:  ' + ValidMigliorClustering +
      '  dall\'indice di validazione rispettivo ottenuto dal dataset random')
```
