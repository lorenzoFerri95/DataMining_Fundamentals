## Association Rules Analysis

```python
rules = apriori(baskets_All_Training,
                supp = 0.01,
                conf = 80,
                zmin=2, zmax=4,
                target='r',
                report='aSCl')
```

```python
print('Number of rules:', len(rules))
```

```python
for r in rules[:3]:
    print(r)
    print('r[0] = -> B = ' + str(r[0]),
          '\nr[1] = A -> = ' + str(r[1]),
          '\nr[2] = Supporto in assoluto (n. elem) = ' + str(r[2]),
          '\nr[3] = Supporto in percentuale = ' + str(r[3]),
          '\nr[4] = Confidence in percentuale = ' + str(r[4]),
          '\nr[5] = Lift in assoluto = ' + str(r[5])
          + '\n')
```

** - Top Rules **

```python
topRules = list()

for r in rules:
    if r[3] <= 5 and r[5] >= 2:
        topRules.append(r)

print('\nNumero di topRules: ' + str(len(topRules)))
```

```python
del rules
```

** - Top Rules con il valore da minare **

```python
topRules_valueToMine = list()

for r in topRules:
    if r[0] == valueToMine:
        topRules_valueToMine.append(r)
        
        print(r)
        
print('\n\nNumero di Top Rules in cui è presente ' + valueToMine + ': ' + str(len(topRules_valueToMine)))
```

```python
del topRules
```

## Predizione del valore minato

```python
valueToMine_Pred = valueToMine + '_Pred'
```

- Training set

```python
for b in baskets_All_Training:
    for r in topRules_valueToMine:
        if (valueToMine_Pred not in b) and (set(r[1]) < set(b)):
            b.append(valueToMine_Pred)
```

- Test Set

```python
for b in baskets_All_Test:
    for r in topRules_valueToMine:
        if (valueToMine_Pred not in b) and (set(r[1]) < set(b)):
            b.append(valueToMine_Pred)
```
