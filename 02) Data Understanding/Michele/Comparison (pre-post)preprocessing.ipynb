# Comparison (pre-post)preprocessing

```python
%matplotlib inline
import math
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

from collections import defaultdict
from scipy.stats.stats import pearsonr
```

```python
pre = pd.read_csv('../../Dataset/training.csv')
post = pd.read_csv('../../Dataset/newTrain.csv')
```

Gli unici attributi che sono stati sottoposti a modifica sono model e submodel

# Model:
passaggio da 1029 a 264 valori distinti <br>
missing values=0%

```python
# Set up a grid of plots
fig = plt.figure(figsize=(20, 20)) 
fig_dims = (3, 2)

# Plot old Distribution
plt.subplot2grid(fig_dims, (0, 0))
pre["Model"].value_counts().plot(kind="bar", title='Old distrib')

# Plot new Distribution
plt.subplot2grid(fig_dims, (0, 1))
post["Model"].value_counts().plot(kind="bar", title='New distrib')

plt.show()
```

```python
pre['Model'].nunique()
```

```python
post['Model'].nunique()
```

```python
(post['Model'].isnull().sum()/post.shape[0])*100
```
