## DBSCAN Perform

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv("Training_DU_Clustering.csv", skipinitialspace=True, sep=',')
df
```

*** Variabili ***

```python
columns_toAnalyze = df.columns
```

```python
M_Values = df.values
```

```python
N_Oggetti = len(df)  # oppure anche:  df.shape[0]
N_Attributi = len(columns_toAnalyze)

print('Numero di Oggetti: ' + str(N_Oggetti), '\n' + 'Numero di Attributi: ' + str(N_Attributi))
```

```python
del df
```

```python
iterazioni = ['primo', 'secondo', 'terzo', 'quarto', 'quinto', 'sesto', 'settimo', 'ottavo', 'nono', 'decimo', 'undicesimo',
             'dodicesimo', 'tredicesimo', 'quattordicesimo', 'quindicesimo', 'sedicesimo', 'diciassettesimo',
             'diciottesimo', 'diciannovesimo', 'ventesimo']
```

```python
k = 5  # Numero minimo di punti entro l'eps per avere un Core Point
```

Dobbiamo scegliere il livello di eps in base al grafico delle distanza dai quinti punti più vicini ad ogni oggetto

```python
dist = pdist(M_Values, 'euclidean')
dist = squareform(dist)
print(dist)
```

```python
kth_distances = list()
for d in dist:
    index_kth_distance = np.argsort(d)[k]
    kth_distances.append(d[index_kth_distance])
```

```python
plt.plot(range(0, len(kth_distances)), sorted(kth_distances))
plt.ylabel('Dist dal Quinto Vicino', fontsize=18)
plt.xlabel('Oggetti Ordinati per Dist', fontsize=18)
plt.tick_params(axis='both', labelsize=10)
plt.axis([0, N_Oggetti + 2000, 0, 3])
plt.grid(color='black', linestyle=':', linewidth=0.5)
plt.show()
```

```python
del dist
del index_kth_distance
del kth_distances
del d
```

```python
eps_level=0  # eps iniziale (viene incrementato di 0,1 ad ogni iterazione)
```

** DBSCAN Perform **

```python
Labels = {}
Silh = {}

for i in iterazioni:
    eps_level = eps_level + 0.1
    dbscan = DBSCAN(eps = eps_level, min_samples = k)
    dbscan.fit(M_Values)
    
    Labels["Labels_{}Clustering".format(i)] = dbscan.labels_
    
    if ((np.unique(Labels["Labels_{}Clustering".format(i)]).shape[0] > 10) or
        (np.unique(Labels["Labels_{}Clustering".format(i)]).shape[0] < 2)):
        continue
    
    Silh["Silh_{}Clustering".format(i)] = silhouette_score(M_Values,
                                                           Labels["Labels_{}Clustering".format(i)])
    
    print(i + ' CLUSTERING \n')
    
    print('MinPoints = 5, EPS = '+ str(eps_level))
    print('Numero di Clusters: ' + str(np.unique(Labels["Labels_{}Clustering".format(i)]).shape[0]))
    print('Numero di Oggetti nei Clusters: ',
          np.unique(Labels["Labels_{}Clustering".format(i)], return_counts=True)[1])
    print("Silh_{}Clustering = ".format(i),
          Silh["Silh_{}Clustering".format(i)], '\n \n')
```

```python
print('LABELS: \n', Labels, '\n \n')
print('SILHOUETTES: \n', Silh)
```

```python
del eps_level
del M_Values
```

## DBSCAN Validation

genero un DataFrame con oggetti random (normal standard) di pari dimensioni del df qui sopra e ne calcolo il silhouette di 10 clusterizzazioni per confrontarlo.

```python
Random_df = pd.DataFrame(np.random.randn(N_Oggetti, N_Attributi), columns = columns_toAnalyze)
Random_df
```

```python
del columns_toAnalyze
```

*** Variabili ***

```python
M_Values_randomDF = Random_df.values
```

```python
dist = pdist(M_Values_randomDF, 'euclidean')
dist = squareform(dist)
print(dist)
```

```python
kth_distances = list()
for d in dist:
    index_kth_distance = np.argsort(d)[k]
    kth_distances.append(d[index_kth_distance])
```

```python
plt.plot(range(0, len(kth_distances)), sorted(kth_distances))
plt.ylabel('Dist dal Quinto Vicino', fontsize=18)
plt.xlabel('Oggetti Ordinati per Dist', fontsize=18)
plt.tick_params(axis='both', labelsize=10)
plt.axis([0, N_Oggetti + 2000, 0, 3])
plt.grid(color='black', linestyle=':', linewidth=0.5)
plt.show()
```

```python
eps_level = 0
```

```python
del dist
del index_kth_distance
del kth_distances
del d
```

** DBSCAN Perform **

```python
Labels_rnd = {}
Silh_rnd = {}

for i in iterazioni:
    eps_level = eps_level + 0.1
    dbscan = DBSCAN(eps = eps_level, min_samples = k)
    dbscan.fit(M_Values_randomDF)
    
    Labels_rnd["Labels_{}RandomClustering".format(i)] = dbscan.labels_
    
    
    if ((np.unique(Labels_rnd["Labels_{}RandomClustering".format(i)]).shape[0] > 10) or
        (np.unique(Labels_rnd["Labels_{}RandomClustering".format(i)]).shape[0] < 2)):
        continue
    
    
    Silh_rnd["Silh_{}RandomClustering".format(i)] = silhouette_score(M_Values_randomDF,
                                                                     Labels_rnd["Labels_{}RandomClustering".format(i)])
    
    print(i + ' RANDOM CLUSTERING \n')
    
    print('MinPoints = 5, EPS = '+ str(eps_level))
    print('Numero di Clusters: ' + str(np.unique(Labels_rnd["Labels_{}RandomClustering".format(i)]).shape[0]))
    print('Numero di Oggetti nei Clusters: ',
          np.unique(Labels_rnd["Labels_{}RandomClustering".format(i)], return_counts=True)[1])
    print("Silh_{}RandomClustering = ".format(i),
          Silh_rnd["Silh_{}RandomClustering".format(i)], '\n \n')
```

```python
print('RANDOM LABELS: \n', Labels_rnd, '\n \n')
print('RANDOM SILHOUETTES: \n', Silh_rnd)
```

```python
del eps_level
del M_Values_randomDF
```

# Considerazioni Finali

la miglior clusterizzazione è quella con indice di validazione più distante dal'indice rispettivo ottenuto da un dataset random.

```python
Valid_Clust = {}

for i in iterazioni:
    try:
        Valid_Clust["Valid_{}Clustering".format(i)] = (Silh["Silh_{}Clustering".format(i)] -
                                                       Silh_rnd["Silh_{}RandomClustering".format(i)])
    except:
        continue

for clust, val in Valid_Clust.items():
    if val == max(Valid_Clust.values()):
        Best_Clust = clust[6:-10]
        print('il miglior Clustering è il ' +
              Best_Clust +
              ' Clustering')
```

```python
print('prime 40 Labels del Best Clust: \n', Labels["Labels_{}Clustering".format(Best_Clust)][:40])
```

```python
print('Numero di Oggetti nei Clusters: ', np.unique(Labels["Labels_{}Clustering".format(Best_Clust)], return_counts=True)[1])
```
