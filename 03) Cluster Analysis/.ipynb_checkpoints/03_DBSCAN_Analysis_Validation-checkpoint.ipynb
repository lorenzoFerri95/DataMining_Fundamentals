## Analisi DBSCAN

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv("Training_DU_Clustering.csv", skipinitialspace=True, sep=',')
df
```

```python
Final_df = pd.read_csv("Training_DU_Clustering_forAnalysis.csv", skipinitialspace=True, sep=',')
Final_df
```

## Primo DataFrame

```python
columns_toAnalyze = ['VehicleAge', 'WheelTypeID', 'VehOdo', 'MMRAcquisitionAuctionAveragePrice', 'VehBCost', 'WarrantyCost',
                    'TransmissionID', 'NationalityID']

primoDF = df[columns_toAnalyze]

primoDF.head()
```

```python
primoDF.head().values
```

Prima di lanciare direttamente l'analisi DBSCAN plottiamo il grafico di tutti i 10135 punti del dataframe ordinati per distanza crescente dal Quinto punto più vicino di ciscuno di essi (le ordinate rappresentano questa distanza) 

```python
M_Values_primoDF = primoDF.values
```

```python
dist = pdist(M_Values_primoDF, 'euclidean')
dist = squareform(dist)
print(dist)
```

```python
k = 5
kth_distances = list()
for d in dist:
    index_kth_distance = np.argsort(d)[k]
    kth_distances.append(d[index_kth_distance])
```

```python
plt.plot(range(0, len(kth_distances)), sorted(kth_distances))
plt.ylabel('dist from %sth neighbor' % k, fontsize=18)
plt.xlabel('sorted distances', fontsize=18)
plt.tick_params(axis='both', labelsize=10)
plt.grid(color='black', linestyle=':', linewidth=0.5)
plt.show()
```

** 1) Primo Clustering del Primo Dataframe**

Parametri:
* **eps**: distanza entro la quale deve trovarsi il numero minimo di punti dato da  min_samples  per poter categorizzare il punto in esame come "Core Point".
* **min_samples**: numero minimo di punti che devono trovarsi entro l'EPS per poter categorizzare il punto in esame come "Core Point".

```python
dbscan = DBSCAN(eps=1, min_samples=k)
dbscan.fit(M_Values_primoDF)
```

Output:
* **labels_**: Cluster labels per ogni punto nel dataset. i Noise Point hanno label -1.

```python
Labels_primoDF_primoClust = dbscan.labels_
```

```python
Labels_primoDF_primoClust[:40]  # prime 40 label dei primi 40 punti
```

Vediamo qual'è la size di ciascun cluster

```python
np.unique(Labels_primoDF_primoClust, return_counts=True)
```

```python
Final_df['labels_primoDF_primoClust'] = Labels_primoDF_primoClust
Final_df
```

** 2) Secondo Clustering del Primo Dataframe**

```python
dbscan = DBSCAN(eps=1.5, min_samples=k)
dbscan.fit(M_Values_primoDF)
```

```python
Labels_primoDF_secondoClust = dbscan.labels_
```

```python
Labels_primoDF_secondoClust[:40]
```

```python
np.unique(Labels_primoDF_secondoClust, return_counts=True)
```

```python
Final_df['Labels_primoDF_secondoClust'] = Labels_primoDF_secondoClust
Final_df
```

```python
dbscan = DBSCAN(eps=2, min_samples=k)
dbscan.fit(M_Values_primoDF)
```
